# Bayesian Neural Networks

Nuestro problema es tomar el problema

$$\frac{\text{d}\vec{x}}{\text{d}t} = F(\vec{x}, t)$$

y aproximar $F$ con una red neuronal $NN_\theta(x)$ con parámetros $\theta$ y un conjunto de datos de entrenamiento $\mathcal{D}$.
Dada una función de costo $\mathcal{L}(\theta; \mathcal{D})$, obtenemos un conjunto de parámetros para la red $\theta^{\star} = \arg \min_\theta(\mathcal{L}(\theta; \mathcal{D}))$. Esta minimización puede interpretarse como una maximización indirecta del _likelihood_ $P(\mathcal{D}\vert \theta)$ y si se introducen términos de regularización en la función de costo (equivalente a introducir _priors_ $P(\theta)$), una maximización indirecta del _posterior_ $P(\theta \vert \mathcal{D})$.

Una desventaja de este enfoque es que al obtener únicamente el máximo de este posterior, se está tirando información dada por el resto de la distribución. Esta distribución es muy importante si se quieren hacer estimaciones probabilísticas de las predicciones $y$ de la red neuronal puesto que 

$$P(y \vert x, \mathcal{D}) = \int \text{d}\theta\hspace{5pt} P(y\vert x,\theta)P(\theta \vert \mathcal{D})$$

Esta integral puede ser aproximada por 

$$\sum_{\theta \sim P(\theta \vert \mathcal{D})}NN_\theta(x)$$

De aquí vemos que el problema se puede reducir a obtener muestras $\theta \sim P(\theta \vert \mathcal{D})$ que se puede hacer con MCMC. Esto es lo que haremos ahora con el modelo Lotka-Volterra

```julia
using Turing, DiffEqFlux, Plots, Flux, Distributions
using OrdinaryDiffEq
```

```julia
#Generación de datos 
function lotka_volterra(du,u,p,t)
  x, y = u
  α, β, δ, γ = p
  du[1] = dx = (α - β*y)x
  du[2] = dy = (δ*x - γ)y
end
p = [1.5, 1.0, 3.0, 1.0]
u0 = [1.0,1.0]
prob = ODEProblem(lotka_volterra,u0,(0.0,10.0),p)
sol = solve(prob,Tsit5())
plot(sol)
```

```julia
odedata = Array(solve(prob,Tsit5(),saveat=0.1))
```

```julia
#nuestros datos de entrenamiento
odedata = odedata[:,1:50]
```

```julia
#una red neuronal muy simple 
ann = FastChain(FastDense(2, 5, tanh), FastDense(5,2))
nn_ode = NeuralODE(ann, (0.0, 4.9), alg = Vern7(),saveat = 0.1)
n_params = length(nn_ode.p)
```

```julia
#Un wrapper para obtener el sampleo (por alguna razón no jala con NeuralODE)
function dudt_node(u, p, t)
    du1, du2 = ann(u, p)
    [du1, du2]
end
```

```julia
#Declaramos nuestro modelo
@model function fitlv(data)
    #priors
    σ ~ InverseGamma(2, 3)
    p ~ MvNormal(zeros(n_params), 0.5)
    
    prob = ODEProblem(dudt_node, u0, (0.0, 5.0), p)
    predict = solve(prob, Tsit5(),saveat = 0.1)
    #likelihood
    for i = 1:size(data,2)
        data[:,i] ~ MvNormal(predict[:,i], σ)
    end
end
```

```julia
#sampleo de nuestros parámetros
model = fitlv(odedata)
chain = sample(model, NUTS(0.65), 500)
```
